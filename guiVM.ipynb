```python
import tkinter as tk
from tkinter import messagebox
from tkinter.filedialog import *
from tkinter import Entry
import os
from tkinter import Button
root=tk.Tk()
root.title('Student Info System')
root.geometry('1300x600+0+0') # it indicate by windows revaluation height or width or starting  point
root.config(bg='black')

#creating Variables
#string
sid=tk.StringVar()
fname=tk.StringVar()
lname=tk.StringVar()
clss=tk.StringVar()
stre=tk.StringVar()
email=tk.StringVar()


#INT
pyt=tk.IntVar()
java=tk.IntVar()
html=tk.IntVar()
total=tk.IntVar()
per1=tk.IntVar()
gd=tk.StringVar()




def save_file():
    with open('data/' + sid.get() + '.txt','w')as f:
        f.write(sid.get()+',' + fname.get() +',' + lname.get()+',' + clss.get()+',' + stre.get()+',' + email.get()+',' + str(pyt.get()) +',' + str(java.get()) +',' + str(html.get()) +',' + str(total.get()) +',' + str(per1.get()) +',' + str(gd.get()))

        
def save():
    present = 'no'
    if sid.get() == '':
         messagebox.showerror('Warning', 'Include Necessary Data')
    else:
        f = os.listdir('data/')
        print(f)
        for file in f:
            if sid.get() == file.split('.')[0]:
                present = 'yes'
        if present == 'yes':
            ask = messagebox.askyesno('Warning', 'Do you really want to save ')
            if ask>0:
                save_file()
                messagebox.showinfo('Sucess','File Update Sucessfully')
                
        else :
            save_file()
            messagebox.showinfo('Sucess', 'File Save Sucessfully')

                    





def clear():
    ser.set('')
    sid.set('')
    fname.set('')
    lname.set('')
    clss.set('')
    stre.set('')
    email.set('')
    pyt.set('')
    java.set('')
    html.set('')
    total.set('')
    per1.set('')
    gd.set('')

    
    
    
def calculate():
    
    
    total1 = pyt.get() + java.get() + html.get()
    total.set(total1)
    per = total1 / 300.0 * 100
    per1.set(per)
    
    if per>=91 and per<=100:
        gd.set('O+')
    elif per>=81 and per<91:
        gd.set('O')
    elif per>=71 and per<81:
        gd.set('A')
    elif per>=61 and per<71:
        gd.set('B')
    elif per>=51 and per<61:
        gd.set('C')
    elif per>=40 and per<51:
        gd.set('D')
    else:
        gd.set('Invalid Input!')


def logout():
    ask = messagebox.askyesno('Warning','Do You Really Want To Logout')
    #ask=tk.messagebox.askyesno('Warning','Do YOu Really Want To Logout')
    print(ask)
    
    if ask>0:
        root.destroy()


# This code for top frame
top_frame=tk.Frame(root,bg='black')
top_frame.place(x=0,y=0,width=1350,height=200)
top_frame_lbl=tk.Label(top_frame,text='Student Info System',font=('time new roman',36,'bold'),fg='grey')
top_frame_lbl.pack(pady=70)



# This code for left Frame
left_frame=tk.Frame(root,bg='gray')
left_frame.place(x=0,y=200,width=675,height=300)
left_frame_info=tk.Label(left_frame,text='Personal Info Of Student. :',font=('time new roman',16,'bold'),bg='gray',fg='black').grid(row=0,columnspan=2)
left_frame_sid=tk.Label(left_frame,text='Student PRN No. :'.ljust(50),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=1,columnspan=2)
left_frame_fname=tk.Label(left_frame,text='First name :'.ljust(26),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=2)
left_frame_lname=tk.Label(left_frame,text='Last name :'.ljust(25),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=3)
left_frame_clss=tk.Label(left_frame,text='Class :'.ljust(30),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=4)
left_frame_stre=tk.Label(left_frame,text='Stream :'.ljust(28),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=5)
left_frame_email=tk.Label(left_frame,text='E-mail ID :'.ljust(28),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=6)


#e1 = Entry(left_frame)
e2 = Entry(left_frame,textvariable=sid) 
e3 = Entry(left_frame,textvariable=fname)
e4 = Entry(left_frame,textvariable=lname)
e5 = Entry(left_frame,textvariable=clss)
e6 = Entry(left_frame,textvariable=stre)
e7 = Entry(left_frame,textvariable=email)




#e1.grid(row=0, column=1)
e2.grid(row=1, column=1) 
e3.grid(row=2, column=1) 
e4.grid(row=3, column=1)
e5.grid(row=4, column=1)
e6.grid(row=5, column=1)
e7.grid(row=6, column=1)



#left_frame_sid.place(x=0,y=0)
#left_frame_fname.place(x=0,y=75)
#left_frame_lname.place(x=0,y=150)
#left_frame_clss.place(x=0,y=200)
#left_frame_stre.place(x=0,y=250)


#left_frame_fname.pack(pady=70)


#This Code for Right frame
right_frame=tk.Frame(root,bg='gray')
right_frame.place(x=675,y=200,width=675,height=300)
right_frame_marks=tk.Label(right_frame,text='Subject Marks',font=('time new roman',16,'bold'),bg='gray',fg='black').grid(row=0,columnspan=2)
right_frame_pyt=tk.Label(right_frame,text='Python :'.ljust(26),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=1)
right_frame_java=tk.Label(right_frame,text='Java :'.ljust(28),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=2)
right_frame_html=tk.Label(right_frame,text='HTML :'.ljust(26),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=3)
right_frame_total=tk.Label(right_frame,text='Total :'.ljust(28),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=4)
right_frame_per=tk.Label(right_frame,text='percentage :'.ljust(20),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=5)
right_frame_gd=tk.Label(right_frame,text='Grade :'.ljust(26),font=('time new roman',12,'bold'),bg='gray',fg='black').grid(row=6)




#e1 = Entry(right_frame) 
e2 = Entry(right_frame,textvariable=pyt)
e3 = Entry(right_frame,textvariable=java)
e4 = Entry(right_frame,textvariable=html)
e5 = Entry(right_frame,textvariable=total)
e6 = Entry(right_frame,textvariable=per1)
e7 = Entry(right_frame,textvariable=gd)



#e1.grid(row=0, column=2) 
e2.grid(row=1, column=2) 
e3.grid(row=2, column=2)
e4.grid(row=3, column=2)
e5.grid(row=4, column=2)
e6.grid(row=5, column=2)
e7.grid(row=6, column=2)



#right_frame_marks.place(x=0,y=10)
#right_frame_pyt.place(x=0,y=50)
#right_frame_java.place(x=0,y=100)
#right_frame_html.place(x=0,y=150)
#right_frame_total.place(x=0,y=200)
#right_frame_per.place(x=0,y=250)



#This Code for Bottom frame

bottom_frame=tk.Frame(root,bg='pink')
bottom_frame.place(x=0,y=500,width=1300,height=100)

save1 = Button(bottom_frame, text = 'SAVE',command=save, width=15, fg ='red') 
save1.pack( side = LEFT)

#saveas = Button(bottom_frame, text = 'SAVE AS',command=save_as,width=15, fg ='red') 
#saveas.pack( side = LEFT)

tota = Button(bottom_frame, text = 'TOTAL',command=calculate,width=15, fg ='red') 
tota.pack( side = RIGHT)

clear = Button(bottom_frame, text = 'CLEAR', command=clear,width=15, fg='brown') 
clear.pack( side =RIGHT ) 

logout = Button(bottom_frame, text ='LOGOUT',command=logout, width=15,fg ='blue')
logout.pack( side = BOTTOM ) 

root.mainloop()
```
